2. What do you think about the students' parental level of education?

```python
import matplotlib.pyplot as plt
data['Score']=data['math score']+data['reading score']+data['writing score']
plt.bar(data['parental level of education'],data['Score'])
plt.title('Parental Education')
plt.xlabel('parental level of education')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()

```

3. Who scores the most on average for math, reading and writing based on● Gender● Test preparation course

```python
avg_gender=data.groupby(['gender'])[['math score','reading score','writing score']].mean()
avg_gender
```

```python
avg_course=data.groupby(['test preparation course'])[['math score','reading score','writing score']].mean()
avg_course
```

5. The management needs your help to give bonus points to the top 25% of
students based on their math score, so how will you help the management
to achieve this.

```python
data_sorted = data.sort_values(by='math score', ascending=False)
total_students = len(data_sorted)
top_25_percent_students = int(total_students * 0.25)


bonus_points = 25
top_25_percent_data = data_sorted.head(top_25_percent_students)
top_25_percent_data['bonus_points'] = bonus_points


data['bonus_points'] = 0
data.loc[top_25_percent_data.index, 'bonus_points'] = bonus_points
```
